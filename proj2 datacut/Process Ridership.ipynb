```python
import csv

with open('ridership.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    line1 = None
    dic = {}
    for row in csv_reader:
        if(line_count == 0):
            line1=row
        else:
            if(row[1] in dic):
                dic[row[1]].append(row)
            else:
                dic[row[1]] = []
                dic[row[1]].append(line1)
                dic[row[1]].append(row)
        line_count += 1
    print(f'Processed {line_count} lines.')
```

```python
for i in dic:
    filename = i.replace(' ','_')
    filename = filename.replace('/','__')
    filename = filename+'.csv'
    with open(filename, 'w') as csvfile:
        csvwriter = csv.writer(csvfile) 
        csvwriter.writerows(dic[i])
```

```python
line1 = None
p = []
with open('sysinfo.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    dic = {}
    for row in csv_reader:
        if(line_count == 0):
            line1=row
            line1 += ['Lat']
            line1 += ['Long']
        else:
            latlon = row[-1].split(',')
            lat = latlon[0][1:len(latlon[0])]
            long = latlon[1][1:len(latlon[1])-1]
            row += [lat]
            row += [long]
            p += [row]
        line_count += 1
    print(f'Processed {line_count} lines.')

filename = 'lonlat.csv'
with open(filename, 'w') as csvfile:
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerow(line1)
    csvwriter.writerows(p)
```
